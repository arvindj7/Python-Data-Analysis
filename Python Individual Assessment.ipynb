##### Content Growth Analysis

```python
#check content type distribution
content_type_counts = df['type'].value_counts() # Count the occurrences of each content type
percentages = (content_type_counts / content_type_counts.sum()) * 100
plt.style.use('dark_background') # using this so the visualization reflects Netflix’s aesthetic with its distinct color theme
ax = content_type_counts.plot(kind='barh', figsize=(18, 7), color=['red', 'blue']) # ax=axes, figsize sets the size of figure
plt.title('Distribution of Content Types on Netflix') # title of the plot
plt.xlabel('Number of Titles') #sets label for x axis
plt.ylabel('Content Type') #sets label for x axis

# Annotates percentages on the bars
for i, value in enumerate(content_type_counts):
    plt.text(value + 10, i, f'{percentages[i]:.1f}%', va='center') # va stands for vertical alignment that aligns the text accorgingly

plt.savefig("Distribution of Content Types on Netflix.png", format="png", dpi = 300) #Exports the visualisation as an image file.
plt.show()
```

Based on the analysis of the distribution of movies and TV shows, it is evident that movies represent the majority of the content on Netflix. While TV shows are present in the catalog, their volume is significantly lower compared to movies.

```python
def Annual_Content_Growth(data):
    # Validating input
    if not isinstance(data, pd.DataFrame):
        raise ValueError("Input data must be a pandas DataFrame")
    if 'release_year' not in data.columns or 'type' not in data.columns:
        raise ValueError("DataFrame must contain 'release_year' and 'type' columns")
    
    data['release_year'] = pd.to_numeric(data['release_year'], errors='coerce') # Ensures release_year is numeric
    annual_content = data.groupby(['release_year', 'type']).size().unstack().fillna(0) # Groups data and calculate yearly content trends

    ax = annual_content.plot(kind='line', figsize=(14, 7), color=['red', 'lime'], linewidth=2, marker='o') #kind sets the type of plot
    ax.set_title('Growth of Content over the Year')
    ax.set_xlabel('Release Year')
    ax.set_ylabel('Number of Contents')
    ax.grid(True, linestyle='--', alpha=0.5) #adds glidlines to a plot if True, linestyle creates dashed lines and alpha sets the transparency
    ax.legend(loc='upper left') #loc sets the placing of legend
    plt.tight_layout() #Automatically adjusts the elements in the graph to make the layout more appealing.
    plt.savefig("Growth of Content over the Year.png", format="png", dpi = 300) 
    plt.show()
    
''' Testing
dj=[1,2,3,4]
Annual_Content_Growth(dj) --- Throws error'''
    
Annual_Content_Growth(df)
```

The above graph shows the spike of production for movies and TV shows. But let's check from year 2000 and seperately for content type.

```python
# Grouping data by release year and type
release_counts = df.groupby(["release_year", "type"]).size().reset_index(name="count") #.reset_index() creates a dataframe.

# Separate data for Movies and TV Shows
movies_data = release_counts[release_counts['type'] == 'Movie']
tvshows_data = release_counts[release_counts['type'] == 'TV Show']

# Filter the data to only include years from 2000 onwards
movies_data_filtered = movies_data[movies_data['release_year'] >= 2000]
tvshows_data_filtered = tvshows_data[tvshows_data['release_year'] >= 2000]

# Create the figure and axes for subplots
fig, ax = plt.subplots(2, 1, figsize=(15, 8))  

# Define custom x-ticks for Movies
movies_xticks = np.arange(2000, movies_data['release_year'].max() + 1, 1)
# Define custom x-ticks for TV Shows
tvshows_xticks = np.arange(2000, tvshows_data['release_year'].max() + 1, 1)

# Plot for Movies
ax[0].plot(movies_data_filtered['release_year'],movies_data_filtered['count'], marker='o', color='red', label='Movies')
ax[0].set_title("Number of Movies Released Each Year (Post 2000)", fontsize=14)
ax[0].set_xlabel("Release Year")
ax[0].set_ylabel("Number of Releases")
ax[0].set_xticks(movies_xticks) #seperate x axes for movies
ax[0].grid(True, linestyle='--', alpha=0.6)
ax[0].legend(loc='upper left') # loc is used for placing of legend 

# Plot for TV Shows
ax[1].plot(tvshows_data_filtered['release_year'], tvshows_data_filtered['count'], marker='o', color='lime', label='TV Shows')
ax[1].set_title("Number of TV Shows Released Each Year (Post 2000)", fontsize=14)
ax[1].set_xlabel("Release Year")
ax[1].set_ylabel("Number of Releases")
ax[1].set_xticks(tvshows_xticks) #seperate x axes for TV Shows
ax[1].grid(True, linestyle='--', alpha=0.6)
ax[1].legend(loc='upper left') # loc is used for placing of legend 
plt.tight_layout()
plt.savefig("Movies and TV Shows Split Plot.png", format="png", dpi = 300)
plt.show()
```

The graph above shows massive increase in production of content, post 2000's: Movies and shows have a proper rise starting somewhere from around the year 2000 but there is an exponential increase in the trends from the 2010s; the increase reflects how diverse streaming services have evolved. 

Movies lead in Overall Releases: Movie releases outstrip TV in annual productions.  It follows the trend from its earlier days, when Netflix mostly focused on acquiring movies and streaming before investing heavily into original series.

Peak Production: Both movies and TV shows reach their peak release counts around 2018-2020. This could be attributed to Netflix's aggressive content production strategy to dominate the streaming market. The decline post-2020 could well be due to  the global impact of the COVID-19 pandemic, which disrupted film and TV production timelines. But movies had a downfall way before in 2018 itself.

TV Shows are Growing Steadily: While movies are leading the volume, the growth rate of TV shows stands significantly high, mainly in the last decade. That signals Netflix is moving toward serialized content, which usually brings long-term viewer engagement.

##### Regional Contributions: Top content producing countries and audience trends by analyzing demographic-specific trends.

```python
df1=df[df['country']!= 'Country Unavailable'] #excludes the handled missing values; if included analysis wont be correct.
top_countries = df1['country'].value_counts().nlargest(10) #.nlargest(10) gives us the top 10 largest values
top_countries.plot(kind='line', color='red', figsize=(14, 7), linewidth=2, marker='o')

plt.title("Top 10 Countries Producing Content")
plt.xlabel("Countries")
plt.ylabel("Values")
plt.grid(True, linestyle='--', alpha=0.6) #adds glidlines to a plot if True, linestyle creates dashed lines and alpha sets the transparency
plt.xticks(ticks=range(len(top_countries)), labels=top_countries.index, rotation=45) #rotates x axis values by given degree
plt.tight_layout()
plt.savefig("Top 10 Countries Producing Content.png", format="png", dpi = 300)
plt.show()
```

The graph above shows the top 10 countries producing content. Now that we have the data of the top players let's analyze demographic-specific trends to understand the targeting of content to kids, teens, and adults.

```python
Age_Ratings = {
    'TV-PG': 'Kids Age 7 to 12',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Kids Age 7 to 12',
    'TV-Y7': 'Kids Age 7 to 12',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'All Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'All Kids',
    'PG': 'Kids Age 7 to 12',
    'G': 'All Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
} # dictionary providing proper age for ratings

df['targeted_ages'] = df['rating'].replace(Age_Ratings) #new column targeted_ages with age ratings
df['targeted_ages'].unique()
```

```python
#Using the same code from previous that filters out rows where the country is 'Country Unavailable'
df1 = df[df['country'] != 'Country Unavailable']

# Get the top 10 countries by content count
top_countries = df1['country'].value_counts().nlargest(10)

# Converting `top_countries` to a DataFrame for further manipulation
top_countries_df = top_countries.reset_index()
top_countries_df.columns = ['country', 'count'] 

# Preparing heatmap data 
df_heatmap = df1[df1['country'].isin(top_countries_df['country'])]
df_heatmap = pd.crosstab(df_heatmap['country'],df_heatmap['targeted_ages'],normalize = "index").T
#creates a heatmap-ready DataFrame by using pandas.crosstab
# crosstab counts the frequency of each category in targeted_ages for every country.
# .T - Transposes the DataFrame so that targeted_ages becomes the rows and country becomes the columns.

fig, ax = plt.subplots(1, 1, figsize=(15, 8))

countriess = ['United States', 'India', 'United Kingdom', 'Canada', 'Japan', 'France', 
                 'South Korea', 'Spain', 'Mexico', 'Egypt'] #This is the easiest way for static data
# since we are working on top 10 countries the data can be changed. 
# for dynamic data use groupby: df.groupby('country')['title_count'].sum().nlargest(10).index

agess = ['All Kids','Kids Age 7 to 12','Teens','Adults']

#Seaborn's heatmap function to create a heatmap visualization of a specific subset of the df_heatmap DataFrame
sns.heatmap(df_heatmap.loc[agess,countriess],square=True, linewidth=2.5,cbar=False,
            annot=True,fmt='1.0%',vmax=.6,vmin=0.05,ax=ax) 
#square=True ensures each cell is displayed as a perfect square
#linewidth for width of line that seperates each cell
#cbar stands for color bar is set to False so the bar will not appear on the right hand side which is appearing in null value map
#annot True for annotations to display values
#fmt specifies format of annotation, vmax and vmin max and min value for heatmap color intensity
#ax specifies axes

ax.spines['top'].set_visible(True)

ax.set_yticklabels(ax.get_yticklabels())
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_ylabel('')    
ax.set_xlabel('')
ax.tick_params(axis=u'both', which=u'both',length=0)
plt.tight_layout()
plt.savefig("Audience trends.png", format="png", dpi = 300)
plt.show()
```

Adults are the dominant viewing audience for Netflix globally, especially in Spain at 84% and Mexico at 77%, reflecting a primary focus on mature content. Teens are also a significant secondary audience, especially in India at 57%, South Korea at 42%, Egypt at 68% and Japan at 38%, driven by youth-focused content. In most countries, though, the content for younger children is minimal: Egypt shows just 1% for "All Kids" and 4% for "Kids Age 12 and Under," which shows a gap in family-friendly offerings. The US shows a balanced age group focus, along with South Korea that evenly balances adults with 44%, teens with 42%, whereas kids' content is scarce.

```python
df1=df[df['country']!= 'Country Unavailable']
top_countries = df1['country'].value_counts().nlargest(10).index

# Filtering the dataset for top 10 countries
df_top_countries = df1[df1['country'].isin(top_countries)]

# Grouping data by country and type, then count the number of Movies and TV Shows
split_data = df_top_countries.groupby(['country', 'type'])['title'].count().unstack().fillna(0)

ax = split_data.plot(kind='line', figsize=(14, 7), color=['red', 'lime'], marker='o')

# Manually setting all x-axis ticks and labels as matplotlib automatically adjusts the ticks to avoid overlapping.
plt.xticks(ticks=range(len(split_data.index)), labels=split_data.index, rotation=45)
plt.title('Movies and TV Shows Split in Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.legend(title='Type')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("Movies and TV Shows Split in Top 10.png", format="png", dpi = 300)
plt.show()
```

Based on the chart displaying the split of movies and TV shows across the top 10 countries on Netflix, the United States leads in movies and TV shows, with the country serving as a major content hub for Netflix. India shows a sharp spike in movies but has fewer TV shows, underlining the gap in episodic content. The UK has balanced distributions in terms of content type production. Smaller markets like Canada, Egypt, and France have low representation, indicative of niche catalogs. Movies are leading in most countries, except for the Japan and South Korea with TV shows leading.

```python
top_5_countries = ['United States', 'India', 'United Kingdom', 'Japan', 'South Korea']

# Grouping data by country and year, then count the number of titles
grouped_data = df.groupby(['country', 'release_year']).size().reset_index(name='count')

# Filtering the grouped data for only the specified countries
filtered_data = grouped_data[grouped_data['country'].isin(top_5_countries)]

fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(14, 10)) 
#change nrows and ncols to change dimension of graph and also incase of more number of countries
ax = ax.flatten()

# Plotting each country in its own subplot
for i, country in enumerate(top_5_countries):
    country_data = filtered_data[filtered_data['country'] == country]
    if not country_data.empty:
        ax[i].plot(country_data['release_year'], country_data['count'], marker='o', label=country, color='red')
        ax[i].set_title(country, fontsize=12)
        ax[i].set_xlabel('Year', fontsize=10)
        ax[i].set_ylabel('Content Count', fontsize=10)
        ax[i].grid(True)
        ax[i].legend(fontsize=8)

# Hide empty subplots if any
for j in range(len(top_5_countries), len(ax)):
    ax[j].axis('off')

plt.suptitle('Growth of Content in Top 5 Producers')
plt.tight_layout()
plt.savefig("Growth of Content in Top 5 Producers.png", format="png", dpi = 300)
plt.show()
```

The graph illustrates the growth of content creation by the top five countries. The United States leads the pack, with consistent growth since the 2000s, peaking around 2018-2020. India follows with a sharp increase in movie production post-2010, peaking in 2019. The United Kingdom shows steady growth, with notable peaks in 2018-2020. Although the overall contribution is lesser, Japan and South Korea depict a big growth post-2000, reflecting the global interest in anime and K-content. All countries depict a decline in production after 2020, likely due to the COVID-19 pandemic, and may be a potential disruption in the content pipeline across regions.

##### Genre Distribution

```python
# import Counter class from the collections module.
from collections import Counter 
# Counter class is used for counting occurrences of elements
# Split and count genres (Kapil, 2019)
genres = df['listed_in'].str.split(', ').sum()
genre_counts = Counter(genres).most_common(10)

# Define a Netflix-inspired color palette
netflix_tudum_palette = ['#E50914', '#00FF00', '#0000FF', '#FFD700', '#800080', '#FFA500'] 

genre_df = pd.DataFrame(genre_counts, columns=['Genre', 'Count'])

# Barplot for genres
plt.figure(figsize=(14, 7))
sns.barplot(x='Count', y='Genre', data=genre_df, palette=netflix_tudum_palette) #sns.barplot() creates a bar plot with x and y values and palette for color
plt.title('Top 10 Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.tight_layout()
plt.savefig("Genre Distribution.png", format="png", dpi = 300)
plt.show()
```

International movies dominated Netflix's content library, reflecting its focus on global audiences. Dramas and comedies remain extremely popular, with high storytelling. International TV shows are growing, showing global appeal for serialized content. Action and documentaries hold moderate representation and serve niche audiences but have growth potential. Family and romance genres remain underrepresented, highlighting further opportunities in the "Children & Family Movies" and "Romantic Movies" categories.

## Conclusion

The project presented a deep analysis of trends related to growth, genre and regional contribution of the content mix, as well as demographic penetration for Netflix. The major issue or challenge was handling missing values or null values. The dataset consists highly specific title-based information, it was certainly not practical or useful for any kind of imputation method using mean or mode. Since there were columns like "director," "country," and "cast" that had critical information in the analysis but contained 5% to 40% null values, deleting such records would affect the analysis. These were then replaced with "Unavailable" so that they could remain useful to the analysis. In the "date_added" column, null values were problematic since new column derivations such as "year_added" and "month_added." required them. Since the percentage of missing values in both "date_added" and "duration" was very negligible, it was best to delete the rows with null values. Although the missing value was handled in country column by replacing the null values by unavailable the country column was very important for our visualisation and in top 10 producer plot “country unaviailable” was visible so we had to use df1=df[df['country']!= 'Country Unavailable'] to exclude that.

For movies and TV split, a problem encountered with visualization using the shared x-axis in a split plot, required using content specific custom x-ticks. 

Also, working on heatmaps to look at demographic splits, there were some issues formatting annotations as percentages. These were resolved by adjusting the formatting arguments - fmt='1.0%’.
